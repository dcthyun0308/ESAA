# **2025 전력사용량 예측 AI 경진대회**

# 1. 주제 및 데이터

건물의 전력사용량 데이터, 기상데이터 등 시공간 정보를 활용하여 전력 사용량을 예측하는 AI 모델 개발

train.csv(=100개 건물들의 2024년 06월 01일부터 2024년 08월 24일까지의 85일간의 데이터, 일시별 기온, 강수량, 풍속, 습도, 일조, 일사 정보 포함, 전력사용량(kWh) 포함)

building_info.csv(100개 건물 정보, 건물 번호, 건물 유형, 연면적, 냉방 면적, 태양광 용량, ESS 저장 용량, PCS 용량)

test.csv(100개 건물들의 2024년 08월 25일부터 2024년 08월 31일까지의 데이터
일시별 기온, 강수량, 풍속, 습도의 예보 정보)

 sample_submission.csv(num_date_time은 건물번호와 시간으로 구성된 ID, 해당 ID에 맞춰 전력사용량 예측값을 answer 컬럼에 기입해야 함)

# 2. 코드 리뷰

# EDA
- 건물마다 사용량의 스케일과 변동성이 크게 다름 -> 개체별 모델링 필요성 확인
- 요일, 시간대, 계절적 패턴이 뚜렷하게 나타남
- 결측치는 특정 건물에 집중되어 있으며, 단순 제거 대비 보간 / 대체 전략이 더 적합

--> 데이터가 하나의 덩어리가 아니라 여러 개의 서로 다른 시계열 집합임.

# 전처리
- 파생변수 생성
- 공휴일 변수 생성
- 이상치 보간 : 짧은 시간에 발생하는 스파이크 데이터(정전 or 기기오류)를 선형보간 방식으로 처리
- 이상치 제거 : 긴 기간에 발생하는 스파이크, 원인을 알 수 없는 평균 이동 제거

- Validation(마지막 일주일 데이터를 제외하고 통계량 계산), Test(Train set의 전체 데이터를 활용하여 통계량 계상)의 분리 규칙 준수 후 통계량 사용

# 모델링
- TabPFN
  - TabPFN은 테이블형 데이터를 위한 사전학습 기반 파운데이션 모델
  - 기존 LightGBM/XGBoost처럼 feature importance 기반의 반복 학습이 아님
  - 이미 대규모 합성 데이터에서 학습된 “패턴”을 즉시 적용하여 빠르게 예측하는 방식

# 3. 차별 점 및 배울 점

# 차별 점
- 건물 처리 방식 : 건물 별로 다른 Feature set, 개별 학습
- XGBoost, LightGBM이 아닌 TabPFN 기반 파운데이션 모델 활용
- 후진 제거법으로 건물형 맞춤형 변수 선택

# 배울 점
- 데이터 구조에 따라 개체 단위 모델링이 성능을 높일 수 있었음.
- 파운데이션 모델 TabPFN 선택
