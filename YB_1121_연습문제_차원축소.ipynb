```python
# 데이터 로드
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784', version=1)
```

```python
# train, test split
X_train = mnist['data'][:60000]
y_train = mnist['target'][:60000]
X_test = mnist['data'][60000:]
y_test = mnist['target'][60000:]
```

* 랜덤포레스트에 적용

```python
# 랜덤포레스트에 적용 (n_estimators=10, random_state=42)
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(n_estimators=10, random_state=42, n_jobs=-1)
forest_clf.fit(X_train, y_train)
y_pred_forest = forest_clf.predict(X_test)
```

```python
# 정확도 결과
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred_forest)
print('랜덤 포레스트 정확도:', accuracy)
```

# 1. PCA

* PCA로 변환한 데이터 세트에 랜덤 포레스트 적용

```python
# PCA 라이브러리 불러오기
from sklearn.decomposition import PCA
```

```python
# PCA 적용 95% : n_components=0.95
pca = PCA(n_components=0.95)
```

```python
# PCA 이후 랜덤포레스트 훈련
X_train_pca = pca.fit_transform(X_train)

X_test_pca = pca.transform(X_test)

forest_clf_pca = RandomForestClassifier(n_estimators=10, random_state=42, n_jobs=-1)
forest_clf_pca.fit(X_train_pca, y_train)
y_pred_forest = forest_clf_pca.predict(X_test_pca)
```

```python
# 정확도 결과 (X_test도 PCA 적용해야합니다.)
accuracy_pca = accuracy_score(y_test, y_pred_forest)

print('PCA 적용 후 랜덤 포레스트 정확도:', accuracy_pca)
```

**관련 문제**

* PCA모델을 와인 데이터셋에 적용하여 차원을 축소하고 그 결과를 그래프로 나타내보세요.

(1) 필요한 모듈 불러오기

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

(2)  load_wine을 사용해 데이터셋을 생성한다.

```python
wine = load_wine()
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
```

```python
wine_df.head()
```

(3) X에 표준 스케일링(Standard scaling)을 적용하여 X_scaled을 생성한다.

```python
scaler = StandardScaler()

wine_scaled = scaler.fit_transform(wine_df)
print(f"X_scaled 데이터의 모양: {wine_scaled.shape}")
```

(4) PCA 모델을 X에 적용하여 2개의 차원으로 추출한다. 단, random_state=1234로 설정한다.

```python
#PCA 모델을 X_scaled에 적용하여 2개의 차원으로 추출한다.
pca = PCA(n_components=2, random_state=1234)
wine_pca = pca.fit_transform(wine_scaled)

print(wine_pca.shape)
```

(5) 추출한 피처를 그래프로 나타내라.

```python
pca_df = pd.DataFrame(data=wine_pca, columns=['Principal Component 1', 'Principal Component 2'])

pca_df['Target'] = wine.target

plt.figure(figsize=(8, 6))
plt.title('PCA 결과: 와인 데이터셋의 2차원 시각화 (Target Class 기준)')

sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Target',
    palette='viridis',
    data=pca_df,
    legend='full',
    alpha=0.8
)

plt.xlabel('Principal Component 1 (주성분 1)')
plt.ylabel('Principal Component 2 (주성분 2)')
plt.grid(True)
plt.show()
```

# 2. LDA

- LDA로 변환한 데이터 세트에 랜덤 포레스트 적용

```python
# LDA 라이브러리 불러오기
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# LDA 적용 : n_components=5
lda = LinearDiscriminantAnalysis(n_components=5)
```

```python
# LDA 이후 랜덤포레스트 훈련
X_train_lda = lda.fit_transform(X_train, y_train)

X_test_lda = lda.transform(X_test)

forest_clf_lda = RandomForestClassifier(n_estimators=10, random_state=42, n_jobs=-1)
forest_clf_lda.fit(X_train_lda, y_train)
y_pred_forest_lda = forest_clf_lda.predict(X_test_lda)
```

```python
# 정확도 결과 (X_test도 LDA 적용해야합니다.)
accuracy_lda = accuracy_score(y_test, y_pred_forest_lda)

print('PCA 적용 후 랜덤 포레스트 정확도:', accuracy_lda)
```
