# 악성 URL 분류 AI 경진대회

# 1. 주제 및 데이터

주제
- 악성 URL 분류 AI 알고리즘 개발

데이터
- train.csv - ID : 6,995,056개의 URL, label(0:정상URL, 1:악성URL)

- sample_submission.csv - ID : 샘플별 고유 ID, probability : 예측한 악성 URL일 확률

# 2. 코드 리뷰

EDA
- 기본 특징 : 길이, 서브도메인 개수가 많을수록 악성 비율 높음
- 특수 문자 : 특수문자 개수가 많을수록, '@' 문자를 포함할 경우 악성 비율 높음
- 텍스트 : 대문자를 포함할 경우, URL이 IP 주소로 구성되어 있을 경우 악성 비율 높음
- 구성요소 별 특징 : 서브도메인에 '-' 문자를 포함하는 경우, path 부분의 길이가 길수록 악성 비율 높음
- 엔트로피 : 악성 URL이 정상 URL보다 약간 더 높은 엔트로피 값을 보이며, 이는 무작위로 생성된 문자열일 가능성이 높다는 의미

전처리
- URL을 분석하여 길이, 문자 조성, 도메인 구조, 피싱 패턴 탐지, 도메인 위험도 통계 등 총 30개 이상의 파생 변수를 생성함으로써 URL 문자열을 고유한 특징 벡터로 변환함.

모델링
- 전처리된 모든 파생 피처를 입력값으로 사용해 X, y 구성

- CatBoost, XGBoost, LightGBM 세 가지 트리 기반 모델 각각 학습

- 각 모델이 test 데이터에 대해 “악성 확률” 예측

- 세 모델의 확률을 가중합하여 Soft Voting Ensemble 생성

- 최종 ensemble 확률을 제출 파일로 저장하여 악성 URL 여부 예측

# 3. 차별 점 및 배울 점

차별 점
- URL을 텍스트가 아니라 구조적 데이터로 해석
- 피싱에서 실제로 사용되는 행위 기반 특징 반영
- TLD/SLD 기준 피싱 발생률을 교차검증 기반으로 생성한 점
- 트리 기반 모델을 단일 모델이 아니라 결합함.

배울 점
- 도메인 지식 + 데이터 기반 특징이 모델 성능을 획기적으로 올림
- Feature Engineering이 모델링보다 더 중요할 때가 많음
- CV 기반 Target Encoding의 활용
- 여러 개의 모델을 단순히 앙상블하는 것이 아니라 가중 합산하는 기법
