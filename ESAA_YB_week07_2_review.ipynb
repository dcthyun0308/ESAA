1. 주제 및 데이터

UBC 난소암 하위 유형 분류 및 이상치 감지(UBC-OCEAN)

난소암 아형을 분류하는 것이 목펴. 20개 이상의 의료 센터에서 수집한 세계에서 가장 방대한 난소암 조직병리학 이미지 데이터 세트를 기반으로 학습된 모델을 구축하게 됨.

sample_submission, test, train 세 개의 데이터 주어짐.

2. 코드 리뷰

EDA
- 이미지마다 확대 배율이 다르면, WSI vs TMA 간 차이 존재했음 -> 데이터 형식 및 해상도 검토 필요
- 클래스 불균형 심하고 이상치 탐지
- 학습 데이터에는 "Other" 아형이 포함되지 않앗고, 테스트에 등장할 수 있는 이상 형태를 모델이 일반화해야 했음.

전처리
- 이미지 기반 접근
- 특징 추출용 사전훈련 모델 사용 : Phikon 활용됨
- MIL 알고리즘 적용 : Chowder 모델 사용함
- 앙상블 훈련 구조 : 15 fold * 50 initializations = 약 750 모델 수준의 반복 학습 및 앙상블
- 이상치 탐지 : 분류 확률의 엔트로피 계산하여 "Other" 아형 가능성 판단

모델링
- 입력 : 슬라이드 이미지 패치 단위 특징
- 학습 모델 : MIL 구조 기반 분류기
- 앙상블 방식 : 다양항 초기화와 fold를 이용한 대규모 앙상블
- 일반화 강화 : 학습 데이터-테스트 데이터 간 도메인 변화에 대응하기 위해 모델 calibration 및 검증 전략 강화됨

3. 차별점 및 배운점
차별점
- 일반적인 이미지 분류 방식 대신 병리 슬라이드 특화된 모델구조 사용
- 대규모 앙상블과 calibration 전략을 통해 테스트 일반화 성능을 극대화
- 이상치 탐지를 위한 확률 엔트로피 기반 접근

배운 점
- 데이터 형식, 배율 등에 대한 이해가 모델 성능에 매우 중요하다.
- 데이터 표현 + 학습 전략 + 앙상블 의 조합이 단순히 모델 크기만 키우는 것보다 중요하다
- 병리 이미지처럼 복잡한 입력일 경우, 패치 기반 처리 및 MIL 구조가 효과적이다.
